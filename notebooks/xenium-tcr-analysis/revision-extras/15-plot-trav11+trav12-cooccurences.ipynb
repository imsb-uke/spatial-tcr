```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns


def make_df() -> pd.DataFrame:
    data = [
        (0, 1, 520),
        (0, 2, 85),
        (0, 3, 17),
        (0, 4, 5),
        (1, 0, 102),
        (1, 1, 15),
        (1, 2, 4),
        (1, 3, 3),
        (2, 0, 5),
        (2, 1, 1),
        (2, 2, 1),
        (3, 2, 1),
    ]
    df = pd.DataFrame(data, columns=["AV1_1", "AV1_2", "Occurrence"])
    df["Pair"] = df.apply(lambda r: f"[{r.AV1_1}. {r.AV1_2}.]", axis=1)
    return df


def plot_bar(df: pd.DataFrame) -> None:
    plt.figure(figsize=(7, 4))
    ax = sns.barplot(data=df, x="Pair", y="Occurrence")
    ax.set_xlabel("Observed counts of [TRAV1-1, TRAV1-2]")
    ax.set_ylabel("Number of cells")
    plt.xticks(rotation=0, ha="center")
    ax.bar_label(ax.containers[0])
    sns.despine(ax=ax)
    plt.tight_layout()
    plt.show()


df = make_df()
plot_bar(df)
```
